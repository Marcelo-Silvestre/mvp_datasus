#2-Criação de tabelas delta a partir de cada arquivo .csv carregado.

```sql
%python
path_raw = "/Volumes/mvp_datasus/bronze/raw_data"
path_delta = "/Volumes/mvp_datasus/bronze/tables_delta"

arquivos = dbutils.fs.ls(path_raw)

for arquivo in arquivos:
    if arquivo.name.endswith(".csv"):
        nome = arquivo.name.replace(".csv", "")

        print(f"CRIANDO TABELA DELTA → {nome}")

        df = spark.read.csv(arquivo.path, header=True, inferSchema=True, sep=";")

        df.write.format("delta").mode("overwrite").save(f"{path_delta}/{nome}")

        print(f"Tabela Delta criada: {nome}")
        print("="*120)
```

#3- Conferindo o schema das tabelas e checagem de valores

```sql
%python
# Notebook: Normalização de schema + correções (try_cast-safe)
from pyspark.sql import functions as F, Row
from pyspark.sql.types import IntegerType, StringType

# Ajuste estes caminhos se necessário
path_delta = "/Volumes/mvp_datasus/bronze/tables_delta"
out_report = "/Volumes/mvp_datasus/bronze/qualidade/correcoes_efetuadas"

# Colunas esperadas
required_numeric_years = ["2020","2021","2022","2023","2024","2025"]
required_cols = ["cod_ibge","unidade_federacao"] + required_numeric_years
numeric_columns = ["cod_ibge"] + required_numeric_years

# função segura que converte (usando try_cast) e garante 0 para valores inválidos
def normalize_int_trycast(df, colname):
    """
    Usa try_cast(`colname` AS INT) para tolerar valores malformados e transforma NULL em 0.
    Retorna DataFrame com a coluna convertida para IntegerType.
    """
    safe_expr = f"try_cast(`{colname}` AS INT)"
    return df.withColumn(colname, F.coalesce(F.expr(safe_expr), F.lit(0)).cast(IntegerType()))

# coletor de logs
logs = []

arquivos = dbutils.fs.ls(path_delta)

for arq in arquivos:
    if not arq.isDir():
        continue

    table_name = arq.name.rstrip("/")
    caminho = arq.path
    print(f"\n\n==== PROCESSANDO TABELA: {table_name} ====")

    try:
        df = spark.read.format("delta").load(caminho)
    except Exception as e:
        print(f"Falha ao abrir {table_name}: {e}")
        logs.append(Row(tabela=table_name, acao="LOAD_FAILED", coluna=None, detalhe=str(e), registros_modificados=None))
        continue

    total = df.count()
    print(f"Linhas: {total}")

    orig_dtypes = dict(df.dtypes)

    # 1) Criar colunas numéricas ausentes (com 0)
    for yr in required_numeric_years:
        if yr not in df.columns:
            df = df.withColumn(yr, F.lit(0).cast(IntegerType()))
            logs.append(Row(tabela=table_name, acao="ADDED_MISSING_COLUMN", coluna=yr, detalhe="Coluna ausente criada com 0", registros_modificados=None))
            print(f"  -> coluna {yr} estava ausente. Criada com 0.")

    # 2) Criar cod_ibge se ausente
    if "cod_ibge" not in df.columns:
        df = df.withColumn("cod_ibge", F.lit(0).cast(IntegerType()))
        logs.append(Row(tabela=table_name, acao="ADDED_MISSING_COLUMN", coluna="cod_ibge", detalhe="cod_ibge ausente criada com 0", registros_modificados=None))
        print("  -> cod_ibge ausente. Criado com 0.")

    # 3) Criar unidade_federacao se ausente (NULL)
    if "unidade_federacao" not in df.columns:
        df = df.withColumn("unidade_federacao", F.lit(None).cast(StringType()))
        logs.append(Row(tabela=table_name, acao="ADDED_MISSING_COLUMN", coluna="unidade_federacao", detalhe="unidade_federacao ausente criada com NULL", registros_modificados=None))
        print("  -> unidade_federacao ausente. Criada com NULL.")

    # 4) Normalizar cod_ibge e anos usando try_cast
    for c in numeric_columns:
        # se coluna não existir (já garantimos cod_ibge e anos, mas checagem de segurança)
        if c not in df.columns:
            df = df.withColumn(c, F.lit(0).cast(IntegerType()))
            logs.append(Row(tabela=table_name, acao="ADDED_MISSING_COLUMN", coluna=c, detalhe="Coluna ausente criada com 0", registros_modificados=None))
            print(f"  -> {c} ausente (criado com 0).")
            continue

        dtype = orig_dtypes.get(c)
        # contar inválidos: se coluna é string -> checar rlike e valores '-', '' e NULL
        invalid_count = None
        try:
            if dtype is not None and ("int" in dtype or "bigint" in dtype or "long" in dtype):
                # se já é numérico, inválidos são apenas NULLs
                invalid_count = df.filter(F.col(c).isNull()).count()
            else:
                # se é string ou outro, considerar '', '-', NULL e padrões não-numéricos como inválidos
                invalid_count = df.filter(
                    (F.col(c).isNull()) |
                    (F.trim(F.col(c)) == "") |
                    (F.col(c) == "-") |
                    (~F.col(c).rlike("^-?\\d+$"))
                ).count()
        except Exception as e:
            # fallback: conta apenas NULLs se algo falhar
            invalid_count = df.filter(F.col(c).isNull()).count()

        # aplicar normalize com try_cast (sempre seguro)
        df = normalize_int_trycast(df, c)

        logs.append(Row(
            tabela=table_name,
            acao="FIXED_VALUES",
            coluna=c,
            detalhe=f"Valores inválidos convertidos para 0 via try_cast + coalesce; cast para INT",
            registros_modificados=int(invalid_count) if invalid_count is not None else None
        ))
        print(f"  -> {c}: {int(invalid_count) if invalid_count is not None else 0} registros inválidos convertidos para 0.")

    # 5) Normalizar unidade_federacao: cast para string (se necessário)
    c = "unidade_federacao"
    dtype_uf = orig_dtypes.get(c)
    if dtype_uf != "string":
        df = df.withColumn(c, F.col(c).cast(StringType()))
        logs.append(Row(tabela=table_name, acao="CAST_TYPE", coluna=c, detalhe=f"Cast de {dtype_uf} para string", registros_modificados=None))
        print(f"  -> unidade_federacao: tipo original {dtype_uf} -> convertido para string.")
    else:
        print("  -> unidade_federacao: já é string; sem alterações de tipo.")

    # 6) garantir coluna cod_ibge NOT NULL logicamente (já transformamos nulos em 0)
    # (Se quiser explicitamente disallow NULLs no metastore, isso depende do catálogo; aqui garantimos valores válidos)

    # 7) Regravar a tabela Delta com overwriteSchema
    try:
        df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").save(caminho)
        logs.append(Row(tabela=table_name, acao="TABLE_SAVED", coluna=None, detalhe=f"Tabela sobrescrita em {caminho} com schema corrigido", registros_modificados=None))
        print(f"  -> Tabela {table_name} regravada (overwriteSchema=true).")
    except Exception as e:
        print(f"  ERRO ao gravar tabela {table_name}: {e}")
        logs.append(Row(tabela=table_name, acao="SAVE_FAILED", coluna=None, detalhe=str(e), registros_modificados=None))

# 8) Gerar DataFrame de relatório, mostrar e salvar
if len(logs) == 0:
    print("Nenhuma tabela processada / nenhum log coletado.")
else:
    df_logs = spark.createDataFrame(logs)
    # padrão: ordena por tabela e ação
    display(df_logs.orderBy("tabela"))
    # salvar relatório para auditoria (sobrescreve)
    df_logs.write.mode("overwrite").format("delta").save(out_report)
    print(f"\nRelatório de correções salvo em: {out_report}")

print("\nFim do processamento.")
```
