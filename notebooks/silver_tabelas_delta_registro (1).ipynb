#1- Copiar as tabelas delta da camada bronze para a silver.

```sql
%python
# Lista das tabelas do schema bronze
tabelas = spark.sql("SHOW TABLES IN mvp_datasus.bronze").collect()

for t in tabelas:
    nome = t["tableName"]

    print(f"\n=== Copiando tabela: {nome} ===")

    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS mvp_datasus.silver.`{nome}`
        DEEP CLONE mvp_datasus.bronze.`{nome}`;
    """)

print("\nFinalizado.")
```
